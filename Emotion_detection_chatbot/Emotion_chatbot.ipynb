```python
import pandas as pd

train = pd.read_csv("train.txt", sep=";", names=["text","emotion"])
test = pd.read_csv("test.txt", sep=";", names=["text","emotion"])
val = pd.read_csv("val.txt", sep=";", names=["text","emotion"])

train.head()
```

```python
X_train = train["text"]
y_train = train["emotion"]

X_test = test["text"]
y_test = test["emotion"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()

X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train_vec, y_train)

print("Model trained successfully!")
```

```python
from sklearn.metrics import accuracy_score

predictions = model.predict(X_test_vec)

accuracy = accuracy_score(y_test, predictions)
print("Model Accuracy:", accuracy)
```

```python
def detect_emotion(text):
    text_vec = vectorizer.transform([text])
    prediction = model.predict(text_vec)
    return prediction[0]

detect_emotion("my boyfriend got me flowers")
```

```python
responses = {
    "joy": "That sounds really positive! I'm glad something good happened. 😊",

    "sadness": "I'm sorry you're feeling this way. If you want to talk about it, I'm here to listen.",

    "anger": "It sounds like something frustrating happened. Taking a moment to pause might help.",

    "fear": "That seems worrying. It's completely normal to feel this way sometimes.",

    "love": "That sounds really meaningful and heartwarming. Relationships like that are special.",

    "surprise": "That sounds unexpected! How did it make you feel?"
}
```

```python
def detect_emotion(text):
    text_vec = vectorizer.transform([text])
    prediction = model.predict(text_vec)
    return prediction[0]
```

```python
# Define the chatbot function
def chatbot(user_input):
    emotion = detect_emotion(user_input)
    reply = responses.get(emotion, "Tell me more about how you're feeling.")
    print("Detected Emotion:", emotion)
    print("AI:", reply)

# Start the interactive chatbot loop
print("Chatbot is running... Type 'exit' or 'quit' to stop.")

while True:
    user_input = input("You: ")  # take user input
    if user_input.lower() in ["exit", "quit"]:  # stop condition
        print("Chatbot session ended.")
        break
    chatbot(user_input)  # call your chatbot function
```
